```python
import pandas as pd
```

```python
df = pd.read_excel('process_details.xlsx')
```

```python
for row in df.iterrows():
    if row[1]['state'] == ')': 
        # remove the row
        df.drop(row[0], inplace=True)
```

```python
# Reas process_details.xlsx and find out different unique values in each column and count of each unique value.

# print(df.head())
# print(df.shape)
# print(df.columns)

# Find out the number of unique values in each column.
print(df.nunique())

# Find out the unique values in each column.
# for col in df.columns:
#     print(f'{col}: {df[col].unique()}')
# print(df['Process'].unique())



```

```python
for col in df.columns:
    print(f'{col}: {df[col].unique()}')
```

```python
# convert all numbers to string for all columns and rows
df = df.astype(str)
```

```python
# print graphs indicating number of entries with each unique value in each column.
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 22})

# make a grid of plots
plt.figure(figsize=(30, 300))

iter = 0
for col in df.columns[ :15]:
    if df[col].nunique() == df.shape[0] or df[col].nunique() == 1:
        continue
    # print(col)
    plt.subplot(15, 1, iter+1)
    iter += 1
    plt.title(col)
    plt.bar(df[col].unique(), df[col].value_counts())
plt.show()

```
